## 指数関数のプロット

### 指数関数のプロット１

式(2.2)

$$
V(t) = V \left( 0 \right) \exp \left( G t \right)
$$

について、$V \left( 0 \right) = 100, G = 0.5$の場合と、$V \left( 0 \right) = 5, G = 1.5$の場合をプロットする。

```python
## パラメータの定義
v01 = 100 ## V(0)
g1 = 0.5 ## G

v02 = 5 ## V(0)
g2 = 1.5 ## G
```

```python
## データの準備
t1 = np.arange(0, 10.1, 0.1) ## 0から10までの0.1刻みの数のリスト
v1 = [] ## 空のリスト

t2 = np.arange(0, 10.1, 0.1) ## 0から10までの0.1刻みの数のリスト
v2 = [] ## 空のリスト

## 各tに対応するv(t)の値の生成
### range(x): 0からxまでの１刻みの整数のリスト
### len(x): リストxの長さ（要素数）
for i in range(len(t1)):
  v1.append(v01*np.exp(g1*t1[i]))

for i in range(len(t2)):
  v2.append(v02*np.exp(g2*t2[i]))
```

```python
## 折れ線グラフでプロット
## 小刻みにプロットすることで滑らかな指数関数の曲線がプロットできる
plt.plot(t1, v1)
plt.show()
```

```python
plt.plot(t2, v2)
plt.show()
```

```python
## 重ねてプロット
plt.plot(t1, v1)
plt.plot(t2, v2)
plt.show()
```

### 指数関数のプロット２

式(2.4)

$$
V(t) = V \left( 0 \right) \exp \left( -D t \right)
$$

について、$V \left( 0 \right) = 100, D = 1.5$の場合と、$V \left( 0 \right) = 5, D = 0.5$の場合をプロットする。

```python
## パラメータの定義
v01 = 100 ## V(0)
d1 = 1.5 ## D

v02 = 5 ## V(0)
d2 = 0.5 ## D
```

```python
## データの準備
t1 = np.arange(0, 10.1, 0.1) ## 0から10までの0.1刻みの数のリスト
v1 = [] ## 空のリスト

t2 = np.arange(0, 10.1, 0.1) ## 0から10までの0.1刻みの数のリスト
v2 = [] ## 空のリスト

## 各tに対応するv(t)の値の生成
for i in range(len(t1)):
  v1.append(v01*np.exp(-d1*t1[i]))

for i in range(len(t2)):
  v2.append(v02*np.exp(-d2*t2[i]))
```

```python
plt.plot(t1, v1)
plt.show()
```

```python
plt.plot(t2, v2)
plt.show()
```

```python
plt.plot(t1, v1)
plt.plot(t2, v2)
plt.show()
```

## 指数関数を対数変換した式のプロット

前節の式(2.4)を対数変換した式(2.5)

$$
\log V \left( t \right) = \log V \left( 0 \right) - D t
$$

について、$V \left( 0 \right) = 100, D = 1.5$の場合と、$V \left( 0 \right) = 5, D = 0.5$の場合をプロットする。

```python
## パラメータの定義
v01 = 100 ## V(0)
d1 = 1.5 ## D

v02 = 5 ## V(0)
d2 = 0.5 ## D
```

```python
## データの準備
t1 = np.arange(0, 10.1, 0.1) ## 0から10までの0.1刻みの数のリスト
v1 = [] ## 空のリスト

t2 = np.arange(0, 10.1, 0.1) ## 0から10までの0.1刻みの数のリスト
v2 = [] ## 空のリスト

## 各tに対応するlog(v(t))の値の生成
### log(V(t)) = log(exp(-D*t))
for i in range(len(t1)):
  v1.append(np.log(v01*np.exp(-d1*t1[i])))

for i in range(len(t2)):
  v2.append(np.log(v02*np.exp(-d2*t2[i])))
```

```python
plt.plot(t1, v1)
plt.show()
```

```python
plt.plot(t2, v2)
plt.show()
```

```python
plt.plot(t1, v1)
plt.plot(t2, v2)
plt.show()
```

## ロジスティック関数のプロット

式(2.7)

$$
\frac{d V \left( t \right)}{d t} = G V \left( t \right) \left( 1 - \frac{V \left( t \right)}{K_{V}} \right)
$$

の解、

$$
V \left( t \right) =
   \frac
   {V \left( 0 \right) K_{V} \exp \left( G t\right)}
   {K_{V} - V \left( 0 \right) + V \left( 0 \right) \exp \left( G t \right)}
$$

について、
- $V \left( 0 \right) = 1, G = 1.5, K_{V} = 100$
- $V \left( 0 \right) = 50, G = 0.5, K_{V} = 100$
- $V \left( 0 \right) = 150, G = 0.5, K_{V} = 100$

をプロットする。

参考：[ロジスティック方程式 \- Wikipedia](https://ja.wikipedia.org/wiki/%E3%83%AD%E3%82%B8%E3%82%B9%E3%83%86%E3%82%A3%E3%83%83%E3%82%AF%E6%96%B9%E7%A8%8B%E5%BC%8F)

```python
## パラメータの定義
v01 = 1; g1 = 1.5; kv1 = 100
v02 = 50; g2 = 0.5; kv2 = 100
v03 = 150; g3 = 0.5; kv3 = 100
```

```python
## データの準備
t1 = np.arange(0, 10.1, 0.1) ## 0から10までの0.1刻みの数のリスト
v1 = [] ## 空のリスト

t2 = np.arange(0, 10.1, 0.1) ## 0から10までの0.1刻みの数のリスト
v2 = [] ## 空のリスト

t3 = np.arange(0, 10.1, 0.1) ## 0から10までの0.1刻みの数のリスト
v3 = [] ## 空のリスト

## 各tに対応するlog(v(t))の値の生成
### log(V(t)) = log(exp(-D*t))
for i in range(len(t1)):
  v1.append((v01*kv1*np.exp(g1*t1[i]))/(kv1 - v01 + v01*np.exp(g1*t1[i])))

for i in range(len(t2)):
  v2.append((v02*kv2*np.exp(g2*t2[i]))/(kv2 - v02 + v02*np.exp(g2*t2[i])))

for i in range(len(t3)):
  v3.append((v03*kv3*np.exp(g3*t3[i]))/(kv3 - v03 + v03*np.exp(g3*t3[i])))
```

```python
plt.plot(t1, v1)
plt.plot(t2, v2)
plt.plot(t3, v3)
plt.show()
```

## プロットをきれいにする

デフォルトのプロットのスタイルを少し変えてみる。
- 色を変える
- 線を変える
- マーカーを変える
- 軸の上限、下限を変える
- レジェンドを加える
- タイトル、軸のラベルを加える

参考：Thumbnail gallery — Matplotlib 2.0.2 documentation https://matplotlib.org/2.0.2/gallery.html

```python
## プロットするデータ
x = [1, 2, 3, 4, 5]
y1 = [2, 5, 1, 3, 5]
y2 = [1, 3, 8, 4, 1]
```

```python
## デフォルト
plt.plot(x, y1)
plt.plot(x, y2)
plt.show()
```

```python
## 色を変える
plt.plot(x, y1, color = "red")
plt.plot(x, y2, color = "#0000ff")
plt.show()
```

```python
## 線の種類を変える
plt.plot(x, y1, linestyle = "dashed", color = "black")
plt.plot(x, y2, linestyle = ":", linewidth = 4, color = "grey")
plt.show()
```

```python
## マーカーの種類を変える
plt.plot(x, y1, marker = "o", linestyle = "none", color = "red")
plt.plot(x, y2, marker = "x", linestyle = "none", color = "#0000ff")
plt.show()
```

```python
## 軸の上限、下限を変える
plt.plot(x, y1, color = "red")
plt.plot(x, y2, color = "#0000ff")
plt.xlim(left = 0, right = 8)
plt.ylim(bottom = 0, top = 10)
plt.show()
```

```python
## レジェンドを加える
plt.plot(x, y1, color = "red", label = "Data1")
plt.plot(x, y2, color = "#0000ff", label = "Data2")
plt.legend()
plt.show()
```

```python
## タイトル、軸のラベルを加える
plt.plot(x, y1, color = "red", label = "Data1")
plt.plot(x, y2, color = "#0000ff", label = "Data2")
plt.title("Title")
plt.xlabel("x axis label")
plt.ylabel("y axis label")
plt.legend()
plt.show()
```

## 図を並べる

複数枚の図を１枚の図に並べる。

https://matplotlib.org/stable/api/figure_api.html

```python
## プロットするデータ
x = [1, 2, 3, 4, 5]
y1 = [2, 5, 1, 3, 5]
y2 = [1, 3, 8, 4, 1]
y3 = [5, 5, 1, 5, 3]
y4 = [6, 1, 3, 1, 4]
```

```python
## プロットの準備
fig = plt.figure()

## 1枚目
ax1 = fig.add_subplot(1, 4, 1) ## 1行4列のプロットの1番目
ax1.plot(x, y1)
ax1.set_title("Panel A")

## 2枚目
ax2 = fig.add_subplot(1, 4, 2) ## 1行4列のプロットの２番目
ax2.plot(x, y2)
ax2.set_title("Panel B")

## 3枚目
ax3 = fig.add_subplot(1, 4, 3) ## 1行4列のプロットの２番目
ax3.plot(x, y3)
ax3.set_title("Panel C")

## 4枚目
ax4 = fig.add_subplot(1, 4, 4) ## 1行4列のプロットの２番目
ax4.plot(x, y4)
ax4.set_title("Panel D")

fig.tight_layout()
fig.show()
```
